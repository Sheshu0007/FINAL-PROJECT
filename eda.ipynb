```python
import matplotlib.pyplot as plt

import pandas as pd
```

```python
df=pd.read_csv('cleaned_ecommerce_data.csv')
```

```python
return_rates = df.groupby(['Product_Category', 'User_Location', 'Payment_Method', 'Shipping_Method'])['Returned'].mean().reset_index()
return_rates.rename(columns={'Returned': 'return_rate'}, inplace=True)
```

# Top return reasons

```python
reason_counts = df[df['Returned'] == 1]['Return_Reason'].value_counts()
print("Top return reasons:\n", reason_counts)
```

# Visualize return rates by category

```python


plt.figure(figsize=(10, 6))
df.groupby('Product_Category')['Returned'].mean().sort_values(ascending=False).plot(
    kind='bar', color='skyblue', edgecolor='black'
)
plt.title('Return Rate by Product Category', fontsize=14, weight='bold')
plt.ylabel('Average Return Rate', fontsize=12)
plt.xlabel('Product Category', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

```

# By payment method

```python
plt.figure(figsize=(8, 5))
df.groupby('Payment_Method')['Returned'].mean().plot(kind='bar', color='green')
plt.title('Return Rate by Payment Method')
plt.show()
```
