```python
import pandas as pd
```

```python
df=pd.read_csv('ecommerce_returns_synthetic_data.csv')
```

```python
df
```

# Handle missing values (e.g., fill missing return_reason with 'No Return')


```python
df['Return_Reason'] = df['Return_Reason'].fillna('No Return')
df['Return_Date'] = df['Return_Date'].fillna(pd.NaT)  # NaT for no return
df['Days_to_Return'] = df['Days_to_Return'].fillna(0)  # 0 for non-returns
```

# Convert dates to datetime

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'], errors='coerce')
df['Return_Date'] = pd.to_datetime(df['Return_Date'], errors='coerce')
```

# Remove duplicates and invalid entries (e.g., returns before order date)

```python
df = df.drop_duplicates()
df = df[df['Return_Date'].isna() | (df['Return_Date'] >= df['Order_Date'])]
```

# Ensure 'returned' is binary (if not, create it)

```python
if 'Returned' not in df.columns:
    df['Returned'] = (df['Return_Status'] == 'Returned').astype(int)
```

```python
df['Days_to_Return_Calc'] = (df['Return_Date'] - df['Order_Date']).dt.days
df['Days_to_Return_Calc'] = df['Days_to_Return_Calc'].fillna(0)
```

```python
df.to_csv('cleaned_ecommerce_data.csv', index=False)
print("Cleaned data shape:", df.shape)
```
